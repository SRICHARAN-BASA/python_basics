**Enumerate() usecases**

```python
languages = ["Python","Java","C++","JavaScript"]
# correct usage of enumerate: returns (index, value)
for index, language in enumerate(languages):
    print(f"{index}:{language}")

# if you want indices starting at 1:
print("\nStarting at 1:")
for index, language in enumerate(languages, start=1):
    print(f"{index}:{language}")
```

**Enumerate() example**

```python
fruits = ["apple","banana","cherry"]
for i , fruit in enumerate(fruits, start=1):
    print(f"{i}:{fruit}")
```

**zip() example**

```python
Employees = ["Deepika","Charan","Navaneeth"]
emp_id = [121,123,134]
designation = ["DS","DE","DEV"]
for emp,em_id,design in zip(Employees,emp_id,designation):
    print(f"Employee {emp} and his id is {em_id}and Designation is {design}")
```
